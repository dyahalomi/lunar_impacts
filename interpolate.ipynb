```python
import numpy as np
import pandas as pd
from scipy.interpolate import griddata
from scipy.interpolate import Rbf
```

```python
def interp_impact_rates(phi,theta):
    df = pd.read_csv("./data/flux_rates_high_fidelity_full.csv")
    theta_list = np.array(df["theta"])
    phi_list = np.array(df["phi"])
    flux = np.array(df["flux"])
    impact_rate = flux
    rbf3 = Rbf(theta_list, phi_list, impact_rate, function="multiquadric", smooth=5)
    impact_rate = rbf3(phi, theta)

    return impact_rate

# phi = 70
# theta=120
# interp_impact_rates(phi,theta)
```

```python
theta_base = np.array([77.0, 0.0, 210.0, 220.0, 270.0, 260.0, 250.0, 356.6, 0.0, 45.0, 40.0, 42.0, 85.0])
phi_base = np.array([-87.4, -89.8, -89.5, -89.2, -88.5, -88.7, -88.2, -87.6, -85.8, -84.9, -86.7, -86.5, -84.4])*-1+90
name_base = ["Faustini Rim A", "Peak Near Shackleton", "Connecting Ridge", "Connecting Ridge Extension", "de Gerlache Rim 1","de Gerlache Rim 2", "de Gerlache-Kocher Massif", "Haworth", "Malapert Massif", "Leibnitz Beta Plateau", "Nobile Rim 1", "Nobile Rim 2", "Amundsen Rim"]
```

```python
impact_base = []
for x in range(len(theta_base)):
    rate = interp_impact_rates(phi_base[x], theta_base[x])
    impact_base.append(rate)
```
