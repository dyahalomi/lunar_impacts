```python
import numpy as np
import matplotlib.pyplot as plt


import matplotlib 
matplotlib.rc('xtick', labelsize=23) 
matplotlib.rc('ytick', labelsize=23) 
```

```python
t_w = 0.48
rho_b = 2.851
sigma = 52
S = 11.43
F_2 = 1
rho_p = 8.9
theta = 0
V = np.arange(10,85)
```

```python
d_c = (3.918 * F_2 *  t_w**(2/3)* S**(1/3) *(sigma/70)**(1/3))/(rho_p**(1/3) * rho_b**(1/9) * (V *np.cos(theta))**(2/3))
```

```python
plt.figure(figsize = [18,9])


plt.plot(V, d_c, 'k', lw = 3)
plt.ylabel("critical diamter [cm]", fontsize = 23)
plt.xlabel("velocity [km/s]", fontsize = 23)
plt.title("Ballistic Limit Curve for Whipple Shield", fontsize = 27)


plt.tight_layout()

plt.savefig('ballistic_curve.pdf')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


# Value where the vertical line is to be added
velocity_value = 72

# Find the corresponding critical diameter for velocity = 72 km/s
critical_diameter = np.interp(velocity_value, V, d_c)

# Plot the ballistic limit curve
plt.figure(figsize=[18, 9])
plt.plot(V, d_c, 'k', lw = 4)

# Add the vertical line at velocity = 72 km/s
plt.axvline(x=velocity_value, color='k', linestyle='--', lw=4)
# Annotate the vertical line with velocity label
plt.text(velocity_value - 0.5, plt.ylim()[1] * 0.5, f'Velocity = {velocity_value} km/s', 
         color='k', fontsize=18, rotation=90, va='center', ha='right')

# Annotate the critical diameter value on the curve
plt.plot(velocity_value, critical_diameter, color='k', zorder=3, marker='o', ms=18)
plt.text(velocity_value + 1, critical_diameter + 0.01, r'$r_{\mathrm{crit}} = $' + f'{critical_diameter:.2f} cm', fontsize=18, color='k')


# Fill areas below the curve in green and above the curve in red
plt.fill_between(V, d_c, plt.ylim()[0], color='green', alpha=0.3, label='Below Curve')
plt.fill_between(V, d_c, plt.ylim()[1], color='red', alpha=0.3, label='Above Curve')



# Labels and title
plt.ylabel("Critical Diameter [cm]", fontsize=23)
plt.xlabel("Velocity [km/s]", fontsize=23)
plt.title("Ballistic Limit Curve for Whipple Shield", fontsize=27)

plt.xlim(10,82)
plt.ylim(.1,.4)

# Layout and save
plt.tight_layout()
#plt.savefig('ballistic_curve.pdf')
plt.show()
```
